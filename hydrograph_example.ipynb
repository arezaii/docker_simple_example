# Example loading and visualizing water data

```python
import pandas as pd

wy_2011 = pd.read_csv('water_data/2011.txt','\t',header=None, names=['date','flow'])
wy_2012 = pd.read_csv('water_data/2012.txt','\t', header=None, names=['date','flow'])
wy_2013 = pd.read_csv('water_data/2013.txt','\t', header=None, names=['date','flow'])

wy_2011['dt64']= pd.to_datetime(wy_2011.date)
wy_2012['dt64']= pd.to_datetime(wy_2012.date)
wy_2013['dt64']= pd.to_datetime(wy_2013.date)
```

```python
wy_2011
```

```python
import matplotlib.pyplot as plt
ax=wy_2011.plot(x='dt64', y='flow', kind='line', figsize=(20,10))
ax=wy_2012.plot(x='dt64', y='flow', kind='line', figsize=(20,10))
ax=wy_2013.plot(x='dt64', y='flow', kind='line', figsize=(20,10))
ax.set_ylabel("CFS")
ax.set_xlabel("Date")
```
